```python
# 做完特徵工程前的所有準備 (與前範例相同)
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression

data_path = './data/'
df_train = pd.read_csv(data_path + 'house_train.csv')

train_Y = np.log1p(df_train['SalePrice'])
df = df_train.drop(['Id', 'SalePrice'] , axis=1)
df.head()
```

```python
#只取 int64, float64 兩種數值型欄位, 存於 num_features 中
num_features = []
for dtype, feature in zip(df.dtypes, df.columns):
    if dtype == 'float64' or dtype == 'int64':
        num_features.append(feature)
print(f'{len(num_features)} Numeric Features : {num_features}\n')
```

```python
train_Y.shape
```

```python
# 削減文字型欄位, 只剩數值型欄位
df = df[num_features]
df = df.fillna(-1)
MMEncoder = MinMaxScaler()
train_num = train_Y.shape[0]
df.head()
```

```python
def ecdf(data):
    xaxis= np.sort(data) #sort the data
    yaxis= np.arange(1,len(data)+1)/len(data)   #create percentages for y axis from 1% to 100%
    return xaxis, yaxis

sns.set()

x, y = ecdf(df['1stFlrSF'])
plt.xlabel('sqft')
plt.xticks(np.arange(0, df['1stFlrSF'].max(), 500))
plt.ylabel('Percentage')
plt.plot(x,y, marker='.', linestyle='none')

# 由此可知大部分的data都是在2500 sqft 以下。
```

```python
# 顯示 1stFlrSF 與目標值的散佈圖
sns.regplot(x = df['1stFlrSF'][:train_num], y=train_Y)
plt.show()

# 做線性迴歸, 觀察分數
train_X = MMEncoder.fit_transform(df)
estimator = LinearRegression()
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```

```python
# use Z score to perform a test
from sklearn.preprocessing import StandardScaler
stand = StandardScaler()

train_X = stand.fit_transform(df)
estimator = LinearRegression()
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```
